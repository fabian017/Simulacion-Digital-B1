1. Crear matriz nxm llena de 0
2.La matriz tenga 10 conejos y 3 lobos
3.Realizar:
  
  *C+C=2C
  
  *L+L=2L

  *C+L=-C

```python
import numpy as np
import random
import matplotlib.pyplot as plt

import time


def Conejo(N,n,m,campo):
  su=0
  while True:
    x=random.randrange(n)
    y=random.randrange(m)
    if campo[x][y]=='x':
      campo[x][y]='C'
      su=su+1
      if su==N:
        break
  
  return campo


def Lobo(L,n,m,campo):
  su=0
  while True:
    x=random.randrange(n)
    y=random.randrange(m)
    if campo[x][y]=='x':
      campo[x][y]='L'
      su=su+1
      if su==L:
        break
  
  return campo
  
```

```python
#dimensiones de la matriz
n=int(input("inicialice las dimensiones del bosque: "))
m=int(input("inicialice las dimensiones del bosque: "))
#conejos
C=int(input("Poblacion inicial de Conejos: "))
#Lobos
L=int(input("Poblacion inicial de Lobos: "))
#se crea la poblacion inicial
world=np.full((n, m), 'x') 

world=Conejo(C,n,m,world)
world=Lobo(L,n,m,world)
world
#while


```
