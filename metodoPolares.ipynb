```python
import math as m
import random as rnd
import numpy as np
import time as t
from cgitb import text
import statistics as stat
import scipy.stats as stat2
import sys

N = 10

Z1 = []
Z2 = []

inicio = t.time()
for i in range(N):
    U1 = rnd.random()
    U2 = rnd.random()
    R = m.sqrt(-2*m.log(U1))
    Th = 2 * m.pi * U2
    X = R * m.cos(Th)
    Y = R * m.sin(Th)
    Z1.append(X)
    Z2.append(Y)
final = t.time()



print(Z1)
print(Z2)
'''print(str(N) + " valores de la primera normal estandar")
print(Z1)
# impresión de la primera normal estandar y luego de la segunda
input("Presione enter para seguir")
print(str(N) + " valores de la segunda normal estandar")
print(Z2)

print("Generadas por el método polar")
input("Presione enter para seguir")'''

categ = round(1 + (m.log(len(Z1))/(m.log(2))))
print("{} categorias".format(categ))

maximo = max(Z1)
print("{} máximo".format(maximo))

obs, bins = np.histogram(Z1, categ)

print(f"obs: {obs}")
print(f"Rangos: {bins}")

probs = 1. * np.arange(len(bins)) / float(len(bins) - 1)
print("The CDF result is-",probs)
P = []

for k in range(1,categ+1):
    cc = probs[k] - probs[k-1]
    P.append(cc)

print(f"Probabilidad P entre intervalos: {P}")
print(f"Tamaño de P : {P}")

PROBS = np.array(P)
esp = PROBS * len(Z1)

print(f"Esperados: {esp}")

print(f"Suma de esp: {sum(esp)}")
print(f"Suma de obs: {sum(obs)}")

est_chi = 0

for k in range(categ):
    est_chi = est_chi + ((obs[k] - esp[k])**2)/esp[k]

result = stat2.chisquare(obs, esp, categ - 1, axis = None )
print(f"chi: {result.statistic}")
print(f"chi manual: {est_chi}")
print(stat2.chi2.cdf(est_chi, categ - 1))
print(stat2.chi2.cdf(result.statistic, categ - 1))

```
